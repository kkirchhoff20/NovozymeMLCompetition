```python
import numpy as np

"""
Convert sequence column into |SVD_Dimension|*Fixed_Seq_Length numeric columns

Try regression of ngrams only
THen original aaindex only
Then SVD only
Then 2d scattering wavelet

Try SVM pairs

"""

"""
Input:  1 string valued column per row. 'AAAAKAAALALLGEAPEVVDIWLPAGWRQPFRVFRERKGDGVLVG'
We want: 10*256 numeric columns per row  SVD_1_1 SVD_2_1 SVD_3_1 ... SVD_10_1, sVD_1_2, ..., SVD_1_256, SVD_2_256, SVD_10_256


Row: SVD1_A SVD1_A  SVD1_A SVD1_A SVD1_L SVD1_A ... SVD_1_G SVD_2

"""


def sequence_str_to_svd_arrays(r):
    arrays = []
    s = r['protein_sequence']
    for c in s:
        arrays.append(svd_aaindex[index_map[c]])
    result = np.hstack(arrays)
    # Result is a single row containing 566 values for each original
    #print(result.shape)
    result = result.reshape((len(s), 10)).transpose()# May not need transpose
    
    seq_len = len(s)
    max_len = 256
    result = np.pad(result, [(0, 0), (0, 256-len(s))], 'constant', constant_values=-1)

    #print(result.mean(axis=0))
    #print(result.cumsum(axis=0).min(axis=0))
    #print(result.cumsum(axis=0).max(axis=0))
    
    # Could divide into blocks of 8, 32, etc, and count number of blocks with values above certain quantiles

    return result.reshape(2560)
```
