```python
from scipy.stats import spearmanr
import lightgbm as lgb
lgb_ranker = lgb.LGBMRanker(label_gain = [i for i in range(y_train.max() + 1)])
query_train = [X_train.shape[0]]
query_test = [X_test.shape[0]]
lgb_ranker.fit(X_train[0:10000, :], y_train.head(10000), group=[10000],
        eval_set=[(X_test, y_test)], eval_group=[query_test],
        early_stopping_rounds=5)

round(spearmanr(y_test, lgb_ranker.predict(X_test)).correlation, 5)
```

```python
lgb_regressor = lgb.LGBMRegressor()
lgb_regressor.fit(X_train, train_df["tm"], eval_set=[(X_test, test_df["tm"])], early_stopping_rounds=5)
y_pred = lgb_regressor.predict(X_test)
y_pred_rank = y_pred.argsort().argsort()
round(spearmanr(y_test, y_pred_rank).correlation, 5)
```

```python
aaindex = aaindex.fillna(0)
aaindex.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale, normalize

norm_index = normalize(aaindex, axis=0)
pca_transform = PCA(n_components=10)

pca_transform.fit(norm_index)
pca_aaindex = pca_transform.transform(X=norm_index)
```

```python
pca_aaindex
```

```python
pca_transform.explained_variance_ratio_.cumsum()
```

```python
index_map = {c:i for i, c in  enumerate(list(aaindex.index)) }
index_map
```

```python
pca_aaindex
```

```python
from kymatio.sklearn import Scattering2D
```

```python
import numpy as np

"""
Convert sequence column into |SVD_Dimension|*Fixed_Seq_Length numeric columns

Try regression of ngrams only
THen original aaindex only
Then SVD only
Then 2d scattering wavelet

Try SVM pairs

"""

"""
Input:  1 string valued column per row. 'AAAAKAAALALLGEAPEVVDIWLPAGWRQPFRVFRERKGDGVLVG'
We want: 10*256 numeric columns per row  SVD_1_1 SVD_2_1 SVD_3_1 ... SVD_10_1, sVD_1_2, ..., SVD_1_256, SVD_2_256, SVD_10_256


Row: SVD1_A SVD1_A  SVD1_A SVD1_A SVD1_L SVD1_A ... SVD_1_G SVD_2

"""
#modify such that pca_aaindex is passed as parameter

def sequence_str_to_svd_arrays(r):
    pca_dim = 10
    arrays = []
    s = r['protein_sequence']
    for c in s:
        arrays.append(pca_aaindex[index_map[c]])
    result = np.hstack(arrays)
    # Result is a single row containing 566 values for each original
    #print(result.shape)
    result = result.reshape((len(s), 10)).transpose()# May not need transpose
    
    seq_len = len(s)
    max_len = 512
    result = np.pad(result, [(0, 0), (0, max_len-len(s))], 'constant', constant_values=-1)

    #print(result.mean(axis=0))
    #print(result.cumsum(axis=0).min(axis=0))
    #print(result.cumsum(axis=0).max(axis=0))
    
    # Could divide into blocks of 8, 32, etc, and count number of blocks with values above certain quantiles

    return result.reshape(max_len*pca_dim)
```

```python
column_names = ['X_' + str(x) for x in range(1,5121)]
df[column_names] = df.apply(sequence_str_to_svd_arrays, axis=1, result_type='expand')
```
